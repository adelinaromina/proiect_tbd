```python
import pandas as pd
import re
```

for movie_titles csv file

```python
rows = []
with open(f'dataset/movie_titles.csv', 'r') as file:
    for line in file:
        line_items = [item.strip() for item in line.split(",")]
        
        if len(line_items) > 3:
            row = line_items[:2] + [", ".join(line_items[2:])]
        else:
            row = line_items
        rows.append(row)
df = pd.DataFrame(rows, columns=['movie_id', 'year_of_release', 'title'])
```

```python
# cleaning rows with null
df = df[df["year_of_release"] != "NULL"]
```

```python
df["movie_id"] = df["movie_id"].astype(int)
df["year_of_release"] = df["year_of_release"].astype(int)
df["title"] = df["title"].astype(str)
```

for the combined data txt files

```python
rows = []
for i in range(1,5):
    with open(f'dataset/combined_data_{i}.txt', 'r') as file:
        for line in file:
            line = line.strip()
            match = re.match(r"^(\d+):$", line)
            if match:
                movie_id = match.group(1)
            else:
                customer_id, rating, date = line.split(',')
                row = (movie_id, customer_id, rating, date)
                rows.append(row)

df = pd.DataFrame(rows, columns=['movie_id', 'customer_id', 'rating', 'date'])
```

```python
df["movie_id"] = pd.to_numeric(df["movie_id"], errors="coerce")
df["customer_id"] = pd.to_numeric(df["customer_id"], errors="coerce")
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["date"] = pd.to_datetime(df["date"], errors="coerce")
```
